## 训练自己的词向量

```python
from gensim.models import Word2Vec

sentences = [
    ["我", "喜欢", "机器学习"],
    ["我", "喜欢", "深度学习"],
    ["机器学习", "很", "有趣"],
    ["深度学习", "很", "有趣"],
    ["量子计算"]   # 只出现一次
]

```

```python
model = Word2Vec(
    sentences,          # 已分词的句子序列
    vector_size=100,    # 词向量维度
    window=5,           # 上下文窗口大小
    min_count=2,        # 最小词频（低于将被忽略）
    sg=1,               # 1: Skip-Gram, 0: CBOW
    workers=4           # 并行训练线程数
)

```

```python
# (3) 保存词向量
model.wv.save_word2vec_format('my_vectors.kv')

```

```python
import torch
import torch.nn as nn
from gensim.models import KeyedVectors
import numpy as np
```

```python
wv = KeyedVectors.load_word2vec_format('./data/word2vec.txt', binary=False)
```

```python
unk_token = '<unk>'
index2word = [unk_token] + wv.index_to_key
word2index = {word: index for index, word in enumerate(index2word)}
```

```python
wv.vector_size
```

```python
len(wv.index_to_key)
```

```python
print(wv)
```

```python
num_embeddings = len(index2word)
embedding_dim = wv.vector_size
embedding_matrix = torch.randn(num_embeddings, embedding_dim)

for index, token in enumerate(index2word):
    if token in wv:
        embedding_matrix[index] = torch.tensor(wv[token])
# 把存在电脑上的wv持久化的文件读到内存里，第一行作为unknown的token，其他部分照搬，然后得到的embedding_matrix就是第一行是<unk>:vector[1x100] 其余还是预训练的数据
```

```python
embedding = nn.Embedding.from_pretrained(embedding_matrix)
embedding
```

```python
text = "我喜欢乘坐地铁"
tokens = jieba.lcut(text)
input_ids = [wv.key_to_index[token] for token in tokens] # 没有处理OOV
input_tensor = torch.tensor(input_ids)
embedding(input_tensor)
```

```python
text = "我喜欢乘坐地铁"
tokens = jieba.lcut(text)
input_ids = [word2index.get(token, word2index[unk_token]) for token in tokens]
input_tensor = torch.tensor(input_ids)
embedding(input_tensor)
```
